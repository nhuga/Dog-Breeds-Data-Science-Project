#### Function that takes only the numbers from a string, and then returns the average between the numbers.
##### Example:

input: 10 to 12 years

output: 11

```python
def get_average_from_str(string):
    numbers = re.findall('[0-9]+', string)
    if len(numbers) > 1:
        return ((int(numbers[0]) + int(numbers[1])) / 2)
    elif len(numbers) == 1:
        return int(numbers[0])
    else:
        return np.nan
```

#### API connection function, to get the missing values.

```python
## API to search missing values by breed name and data type
def get_data_from_API(dtype,breed_name):
    # API Request
    API_TOKEN = 'b7eefab0-fa10-4112-b302-88c2e5716616'
    headers = { 'x-api-key': API_TOKEN }
    queryParams = { 'q': breed_name }

    response = requests.get('https://api.thedogapi.com/v1/breeds/search',params = queryParams)
    
    # return specific missing data
    if dtype == 'weight' or dtype == 'height':
        try:
            json_data = json.loads(response.text)[0]
        except:
            print('Not Found!')
        # not found

        try:
            return json_data[dtype]['imperial']
            # '22 - 40'
        except:
            return np.nan
        
    else:
        try:
            json_data = json.loads(response.text)[0]
            return json_data[dtype]
            # '10 - 12 years'
            
        except:
            return np.nan
```

```python
# Fill arr of breed names
def fill_arr_namesbreed(dframe,array):
    for breed in dframe:
        array.append(breed)
```

#### Fill in the missing values by using the get_data_from_API function.

```python
# Call to API get data and update the NaN values in DataFrame 
def update_dataframe_values(breed_arr,dtype,dframe):
    for breed in breed_arr:
        
        try:
            if dtype == 'Weight':
                value = get_data_from_API('weight',breed)
                
            elif dtype == 'Height':
                value = get_data_from_API('height',breed)
                
            elif dtype == 'Life Span':
                value = get_data_from_API('life_span',breed)
                
        except:
            continue
            
        if value:
            dframe.loc[dframe['breed name'] == breed, dtype] = value
```

```python
# fill null values with API

breed_weight_null = df[df['Weight'].isnull() == True]['breed name'];
breed_height_null = df[df['Height'].isnull() == True]['breed name'];
breed_Life_Span_null = df[df['Life Span'].isnull() == True]['breed name'];

# build arr of name breed with missing values
arr_weight = []
arr_height = []
arr_lifespan = []

# fill arr of name breed with missing values
fill_arr_namesbreed(breed_weight_null,arr_weight)
fill_arr_namesbreed(breed_height_null,arr_height)
fill_arr_namesbreed(breed_Life_Span_null,arr_lifespan)

# update dataframe with new data
update_dataframe_values(arr_weight,'Weight',df)
update_dataframe_values(arr_height,'Height',df)
update_dataframe_values(arr_lifespan,'Life Span',df)
```

After we filled in all the missing data values, we fill in the Avg. columns we created above by get_average_from_str function.

```python
df['Weight Avg.'] = df['Weight'].apply(lambda row: get_average_from_str(str(row)))
df['Height Avg.'] = df['Height'].apply(lambda row: get_average_from_str(str(row)))
df['Life Span Avg.'] = df['Life Span'].apply(lambda row: get_average_from_str(str(row)))
df
```

Handling the rest missing data, by fill in the average values.

```python
# Handling missing data
df['Tendency To Bark Or Howl'].fillna(3,inplace = True)
df['Prey Drive'].fillna(3,inplace = True)
df['Drooling Potential'].fillna(3,inplace = True)
df['Weight Avg.'].fillna(df['Weight Avg.'].mean(),inplace = True)
df['Height Avg.'].fillna(df['Height Avg.'].mean(),inplace = True)

df
```
