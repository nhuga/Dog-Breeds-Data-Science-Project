```python
breed_list = breeds['Dog Breed Group'].tolist()
breed_set = set(breed_list)
breed_count = []

for val in breed_set:
    breed_count.append(breed_list.count(val))

breed_count
breed_set
```

```python
for val in breed_set:
    print(val + ' ')
    print(breed_list.count(val))
```

```python
fig, axes = plt.subplots(figsize=(20,8))
pieData = pd.Series(breed_count, index= breed_set)
pieData.plot(kind='pie', ax=axes,autopct='%.2f%%')
axes.set_title("Dog Breed Group in dataset")
axes.set_ylabel("")
```

```python
breeds.info()
```

```python
ct1 = pd.crosstab(breeds['Dog Breed Group'],breeds['Intelligence'])
ct1
```

```python
ctab_Intelligence = pd.crosstab(breeds['Dog Breed Group'],breeds['Intelligence'],normalize=True)
ctab_Intelligence
```

```python
ctab_Intelligence.plot(kind='bar',figsize=(15,5))
plt.title('The level of intelligence in each group',fontsize=30)
plt.xlabel('Dog Breed Group')
plt.ylabel('Intelligence')
```

```python
ctab_Dog_Friendly = pd.crosstab(breeds['Dog Breed Group'],breeds['Dog Friendly'],normalize=True)
ctab_Dog_Friendly
ctab_Dog_Friendly.plot(kind='line',figsize=(15,5))
plt.title('The level of Dog Friendly in each group',fontsize=30)
plt.xlabel('Dog Breed Group')
plt.ylabel('Dog Friendly')
```

```python
fig = plt.figure(figsize=(10,7))
ax = plt.axes()

color_dict = dict({'Mixed Breed Dogs':'brown',
                  'Hound Dogs':'green',
                  'Working Dogs': 'orange',
                  'Hybrid Dogs': 'red',
                   'Companion Dogs': 'dodgerblue',
                  'Herding Dogs':'grey',
                  'Terrier Dogs':'navy',
                  'Sporting Dogs':'salmon'})

sns.scatterplot(x= 'Weight', y = 'Height',s=70,hue=breeds['Dog Breed Group'],palette=color_dict,data={ 'Height': breeds['Height Avg.'] , 'Weight': breeds['Weight Avg.']})
plt.title('Height & Weight by Dog breed group',fontsize=20)
```

#### The dominant feature

```python
clean_df = breeds.copy()
clean_df.drop(['breed name','Height','Affectionate With Family','Height Avg.','Weight','Weight Avg.','Life Span','Life Span Avg.'],inplace=True,axis=1)

listb = list(breed_set)

maximum = 0
feature = ''
for i in range(8):
    for col in clean_df:
        if(col!= 'Dog Breed Group'):
            if (clean_df[clean_df['Dog Breed Group'] == listb[i]].mean()[col] > maximum):
                maximum = clean_df[clean_df['Dog Breed Group'] == listb[i]].mean()[col]
                feature = col
    print("Group name #{name}".format(name = listb[i]))
    print(feature)
    print(maximum)
```
